```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('result.xlsx')
```

```python
# Plot the data
plt.figure(figsize=(10, 6))
for partition in data['pool_size'].unique():
    if not partition or str(partition) == 'nan':
        continue
    partition_data = data[data['pool_size'] == partition]
    plt.plot(partition_data['fine_tuning_epoch'], partition_data['accuracy'], label=f'Client {int(partition)}')

# Customize the plot
# set maximum y-axis value to 1
plt.ylim(0, 100)
# show x as a label
xticks = data['fine_tuning_epoch'].unique()
# remove NaN
xticks = xticks[~pd.isnull(xticks)]
plt.xticks(xticks)
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Accuracy vs Epoch for each Partition')
plt.legend(title='Partition')
plt.grid(True)

# Show the plot
plt.show()
```
